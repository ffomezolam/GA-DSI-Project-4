```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('../data/data-clean-20220916.01.csv')
```

```python
df['Start_Time']= pd.to_datetime(df['Start_Time'])
```

```python
df['day'] = df['Start_Time'].dt.day
df['month'] = df['Start_Time'].dt.month
df['year'] = df['Start_Time'].dt.year
df['hour'] = df['Start_Time'].dt.hour
```

```python
df.head()
```

```python
df = pd.concat([df, pd.get_dummies(df['State'], prefix='State')], axis=1)
df = pd.concat([df, pd.get_dummies(df['Side'], prefix='Side')], axis=1)
df = pd.concat([df, pd.get_dummies(df['Timezone'], prefix='Timezone')], axis=1)
df = pd.concat([df, pd.get_dummies(df['year'], prefix='year')], axis=1)
df = pd.concat([df, pd.get_dummies(df['month'], prefix='month')], axis=1)
df = pd.concat([df, pd.get_dummies(df['hour'], prefix='hour')], axis=1)

#drop the columns: State, Side, Timezone, year, month, hour
df = df.drop(['State', 'Side', 'Timezone', 'year', 'month', 'hour'], axis=1)
```
